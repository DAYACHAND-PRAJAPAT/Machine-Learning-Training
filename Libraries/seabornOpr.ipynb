```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips") 
df.head()
```

```python
import pandas as pd
df = pd.read_csv('tips.csv')
print(df)

df.to_csv("tips.csv",index=False)

sns.lineplot(df["total_bill"])
sns.scatterplot(df["total_bill"])

sns.scatterplot(x=df["total_bill"],y=df["tip"],data=df)

plt.title("graph",fontsize=20,
        color="red",
        fontweight="bold",
        backgroundcolor="yellow",
        style="italic")
```

```python
sns.histplot(df["total_bill"])
```

```python
sns.barplot(x="day",y="total_bill",data=df)
```

```python
sns.kdeplot(df["total_bill"])
```

```python
#sns
##categorical plot --> catplot
##relationship plot --> replot
sns.relplot(x=df["total_bill"],y=df["tip"],data=df)
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="box") #hue for column
```

```python
##outliers --reove ya transform
q1=df["total_bill"].quantile(0.25)
q3=df["total_bill"].quantile(0.75)
ipr=q3-q1
ipr

lower_bound=q1-1.5*ipr
upper_bound=q3+1.5*ipr
df=df[(df["total_bill"]>lower_bound)&(df["total_bill"]<upper_bound)]
sns.boxplot(df["total_bill"])
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="violin")
```

```python
sns.catplot(x="day",y="total_bill",data=df,hue="smoker",kind="point")
```

```python
sns.displot(df["total_bill"],kde=True)
```

```python
sns.jointplot(x="total_bill",y="tip",data=df)
```
